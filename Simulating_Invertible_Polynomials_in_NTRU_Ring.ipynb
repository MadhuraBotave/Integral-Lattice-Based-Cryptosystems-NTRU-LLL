```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

def genPolys(N, p):
    return [np.array(coeffs) for coeffs in np.ndindex(*([p] * N))]

def gcdPolys(f, g, p):
    f_poly = sp.Poly(f, sp.Symbol('x'), modulus=p)
    g_poly = sp.Poly(g, sp.Symbol('x'), modulus=p)
    gcd = sp.gcd(f_poly, g_poly)
    return gcd

def isInver(f, N, p):
    x = sp.Symbol('x')
    g = sp.Poly(x**N - 1, x, modulus=p)
    gcd = gcdPolys(f, g, p)
    return gcd.as_expr() == 1

def simulate_and_plot(p, max_N):
    invertible_counts = []
    total_counts = []

    for N in range(1, max_N + 1):
        invertible_count = 0
        polynomials = genPolys(N, p)
        total_count = len(polynomials)
        total_counts.append(total_count)

        for polyCoeff in polynomials:
            x = sp.Symbol('x')
            f = sp.Poly(polyCoeff, x, modulus=p)
            if isInver(f, N, p):
                invertible_count += 1

        invertible_counts.append(invertible_count)
        print(f'For N={N}, there are {invertible_count} invertible polynomials.')

    # Plotting the bar chart
    N_values = range(1, max_N + 1)
    plt.figure(figsize=(12, 6))
    plt.bar(N_values, invertible_counts, label='Invertible Polynomials', color='b')
    plt.bar(N_values, total_counts, alpha=0.5, label='Total Polynomials', color='r')
    plt.xlabel('N (Degree of Polynomials)')
    plt.ylabel('Number of Polynomials')
    plt.title('Invertible Polynomials vs. Total Polynomials for Increasing N')
    plt.legend()
    plt.show()

p = 3
simulate_and_plot(p, 11)
```
